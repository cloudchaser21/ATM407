### Background

The sense of the QG omega equation, if second derivatives are treated like fancy negative signs and coefficients like static stability are treated as constants, is: 

** (Upward motion) = (Vorticity advection increasing w/ height) + (Warm advection) **

Typically, low-level vorticity advection is weak, so that one can simply interpret *upper level (like 500 or 200 hPa) vorticity advection* as an indicator of the first term. 


Notice that the QG omega equation is a ***diagnostic*** equation, not a **prognostic** equation with a time derivative in it. That is, the right-hand terms do not "lead to" upward motion, they are *equal to it, right now*. 

How did this equation come from our governing Primitive Equations set, which has a prognostic form: *d/dt( {winds, and thickness-related T} )  = {PGF&Coriolis, and physical heating rate Q} )*?

The physical concept behind the QG omega equation is, somehow, to state or assume or express the condition that *the flow remains balanced at all times* -- in other words, $\partial (imbalance)/ \partial t$ = 0. That is, the QG "(Upward motion)" in the equation above is *the hypothetical vertical motion field that exactly compensates the balance-destroying tendencies* of the terms on the right-hand side

## Exercise 0 answer: 


The QG momentum equation is fundamentally a result of influxes of vorticity and temperature into a given region as a result of the geostrophic flow adjacent to the region in question. This flux will inherently create some sort of imbalance that then has to be compensated for via outside forces. For instance, in the case of heterogenous warm air advection into a a region, winds will respond accordingly by changing speed and direction to retain both geostrophic and thermal wind balance. However, the changes that do occur in the horizontal plane actually act to disrupt the balances further (??). As a result some sort of secondary horizontal/vertical circulation must result to compensate for this destructive tendency. This secondary circulation results in upward and downward motion on opposite sides of the newly formed temperature gradient in the region that acts to decrease the overall temperature gradient that arose from the heterogenous temperature advection discussed above. The horizontal components of the secondary circulation that arise from continuity will result in an increase in wind shear as the accompanying coriolis force components act in opposite directions, contributing to the vertical wind shear that has resulted from thermal wind balance. In summary, the QG-w equation works twofold (decreasing thermal gradient and increasing vertical wind shear) to reestablish balance that was destroyed via the introduction of temperature/vorticity imbalances.

## Exercise 7 answer: 

The first animation depicts a coupled upper level trough (denoted by black 500 hPa geopotential height contours) and surface low pressure system (denoted by red MSLP contours) as it sweeps across the northeast. It exhibits what we know to be ideal for surface development: that is, the surface feature is leading the upper level disturbance to some degree. This enables the development of a feedback loop which acts to strengthen the surface disturbance via upper level responses that will then act to deepen the upper level disturbance as a result of surface responses. Leading the upper level trough allows for the surface disturbance to be collocated with the upper level region of greatest 1) positive vertical motion (as a result of 2+3), 2) positive vorticity advection, and 3) warm air advection. Figures 2 and 3, respectively, depict this collocation. Figure 2 shows the region of positive vorticity advection (red shading) relatively collocated with the center of lowest surface pressure. Figure 3 depicts the region of warm air advection (again, red shading) slightly off to the east of the surface low pressure center. As a result of this, we see a large swath of negative omega (referring to positive vertical motions) existing collocated with the surface low pressure center and extending southward into the cold sector (figure 4). This is corroborated by the presence of Q vector convergence in a similar area, a sign of positive rising motions (figures 5+6). Lastly, the optimal upper level/surface disturbance orientation allows for the upper level disturbance to also strengthen (IE deepen) over time as well. This is seen in figure 7, which depicts the time evolution of 500 hPa geopotential heights and omega (vertical motions). In the last frame of the animation in figure 7, we see the upper level trough acquire a completely closed geopotential height contour, indicative of a deepening/intensifying system. This occurs at the same time that positive omega values (IE negative vertical motions) develop in the general area. This further corroborates the notion that cool air advection (depicted in figure 8) is resulting in sinking air and sinking geopotential heights which lead to deepening of the upper level system. This is occuring at the same time as the surface low pressure system acquires tighter pressure contours and establishes a more "comma-like" structure (both indicative of maturation) which does a good job of depicting just how connected surface and upper level development truly is.
