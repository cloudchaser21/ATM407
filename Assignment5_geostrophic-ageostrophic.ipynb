### Background and math

You will decompose the total wind into geostrophic and ageostrophic aspects: $\vec V = \vec V_g + \vec V_a$. 

* Horizontal acceleration equations (F=ma) for inviscid flow:
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv_g $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu_g $$

* The geostrophic wind is defined as *the wind that would occur if the LHS were zero*: 

$$ 0 = - \frac{\partial \Phi}{\partial x} + fv_g $$

$$ 0 = - \frac{\partial \Phi}{\partial y} - fu_g $$

## Exercise 1 Answer:

* Divergence equal to:

$$ \frac{\partial u_g}{\partial x} + \frac{\partial v_g}{\partial y} = \frac{\partial}{\partial x}(-\frac{1}{f}\frac{\partial \Phi}{\partial y}) + \frac{\partial}{\partial y}(\frac{1}{f}\frac{\partial \Phi}{\partial x})$$  
$$ \frac{\partial}{\partial x}(-\frac{1}{f}\frac{\partial \Phi}{\partial y}) + \frac{\partial}{\partial y}(\frac{1}{f}\frac{\partial \Phi}{\partial x}) = -\frac{\partial^2 \Phi}{\partial x\partial y} + \frac{\partial^2 \Phi}{\partial x\partial y} + \frac{\partial}{\partial y}(\frac{1}{f}) = \frac{\partial}{\partial y}(\frac{1}{f})$$ 
* ($f$ doesn't vary with x, double partials cancel)

* if $\beta = df/dy =0$ then $$\frac{\partial}{\partial y}(\frac{1}{f}) = 0$$ 
* thus divergence of the geostrophic wind is equal to 0

#### Key points: the geostrophic wind is boring for 2 reasons: because it is not connected to vertical velocity $\omega$, and also because it would never change with time since its definition sets d/dt to zero. The implication is that ageostrophic wind, despite its relatively small magnitude, is especially *interesting*. However, the geostrophic wind is adequate to approximate the direction and speed of advection of various scalar fields (temperature, vorticity) in the atmosphere.

## Answer to exercise 3: Acceleration-related ageostrophic winds 

In this case, acceleration in the parcels located in or near the region of strong ageostrophic flow is defined as both a change in direction with time and a change in speed with time. At this point in time, the instantaneous acceleration vectors for the true wind such as those located in the square of highest magnitude ageostrophic wind (east of Hawaii) are generally point NNW or in the direction perpendicular to the lowest geopotential height contour (smallest central circle). We know that the acceleration is equal to the coriolis force acting on the ageostrophic flow (in the "j" direction, or perpendicular and to the right of the ageostrophic vectors). Thus, we see that the ageostrophic coriolis vector is equal to and in the same direction of the true wind acceleration vector.
